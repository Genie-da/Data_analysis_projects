## Предобработка данных

### Удаление пропусков

Выведем количество пропущенных значений для каждого столбца.

```python
data.isna().sum()
```

В двух столбцах есть пропущенные значения. Один из них — `days_employed`. Пропуски в этом столбце обработаем на следующем этапе. Другой столбец с пропущенными значениями — `total_income` — хранит данные о доходах. На сумму дохода сильнее всего влияет тип занятости, поэтому заполнить пропуски в этом столбце нужно медианным значением по каждому типу из столбца `income_type`.

```python
# заполним пропуски в total_income медианным значением по каждому типу занятости
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['total_income'].isna()), 'total_income'] = \
    data.loc[(data['income_type'] == t), 'total_income'].median()
```

###### Вывод 
В исходных данных встречаются пропуски в равном количестве в двух столбцах: `days_employed`, отражающий общий трудовой стаж в днях и `total_income`, ежемесячный доход. Это могло бы быть связано с тем, что заемщик не состоял в трудовых отношениях и соответственно не имел никакого дохода, к примеру, студенты, безработные, пенсионеры. Но такая гипотеза противоречит политике банка, и подобным социально незащищенным категориям граждан маловероятно одобрение на кредитование.

Клиенты могли не предоставить данные. Возможно, эту информацию не требуют в некоторых случаях. Также проблемы могли возникнуть на этапе загрузки и на этапе выгрузки данных. Неправильный запрос, объединение таблиц могут быть причиной пропущенных значений. Таким образом, вероятнее всего пропуски связаны с человеческим фактором, в случае предоставления кредитополучателем неполной информации, либо с технологическими причинами.

### Категоризация данных

На основании диапазонов, указанных ниже, создим в датафрейме `data` столбец `total_income_category` с категориями:

- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.

Напишем функцию, которая на основании данных из столбца `income` сформирует новый столбец `total_income_category` с категориями по доходу.

```python
# напишем функцию categorize_income(income), которая принимает в качестве параметра доход и возвращает значение категории
def categorize_income(income):
    try:
        if 0 <= income <= 30000:
            return 'E'
        elif 30001 <= income <= 50000:
            return 'D'
        elif 50001 <= income <= 200000:
            return 'C'
        elif 200001 <= income <= 1000000:
            return 'B'
        elif income >= 1000001:
            return 'A'
    except:
        pass
```

```python
# создадим в датафрейме data столбец total_income_category с категориями по доходу с использованием метода apply()
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

Выведем на экран перечень уникальных целей взятия кредита из столбца `purpose`.

```python
data['purpose'].unique()
```

По данным видно, что присутствуют в основном 4 типа целей. Создадим функцию, которая на основании данных из столбца `purpose` сформирует новый столбец `purpose_category`, в который войдут следующие категории:

- `'операции с автомобилем'`,
- `'операции с недвижимостью'`,
- `'проведение свадьбы'`,
- `'получение образования'`.


```python
# напишем функцию categorize_purpose(), которая принимает в качестве параметра подстроку в столбце purpose
# и возвращает название категории по цели кредита
def categorize_purpose(row):
    try:
        if 'автом' in row:
            return 'операции с автомобилем'
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'
        elif 'свад' in row:
            return 'проведение свадьбы'
        elif 'образов' in row:
            return 'получение образования'
    except:
        return 'нет категории'
```

```python
# создим в датафрейме data столбец purpose_category с категориями по целям кредита с использованием метода apply()
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

###### Вывод 
Для более удобного и точного анализа применили категоризацию данных в столбцах `total_income` и `purpose`. Значения с доходом разделили на 5 категорий: A, B, C, D, E, а цели кредита поделили на 4 типа: операции с автомобилем, операции с недвижимостью, проведение свадьбы, получение образования. Использование категорий улучшает четкость и детализацию данных статистики.

Таким образом, пропуски в исходных данных заполнены, аномальные значениия обработаны, изменен тип данных, проведена обработка дубликитов и категоризация данных, следовательно, этап предобработки данных завершен, приступаем к проверке гипотез.

## Исследование и анализ данных

Для выявления и исследования взаимосвязи надежности заемщиков с такими показателями как: количество детей, семейное положение, уровень дохода и цель получения кредита напишем функцию, которая будет принимать в качестве параметра один из вышеперечисленных показателей и выдавать сводную таблицу по данному показателю.

```python
# Создадим сводную таблицу pivot_table(), где отобразим общее количество заемщиков в general, количество должников в target 
def pivot_table(col):
    df = data.pivot_table(index=[col], values='debt', aggfunc={'count', 'sum'})
    df = df.rename(columns={'count':'general','sum':'target'})
    # Добавим к таблице столбец ratio с соотношением должников к общему числу клиентов, округлим результат до 3 зн после запятой
    df['ratio'] = round(df['target'] / df['general'], 3)
    # Отсортируем таблицу в убывающем порядке по столбцу ratio
    df = df.sort_values(by='ratio', ascending=False)
    return df
```

```python
# Построим сводную таблицу по количеству детей children_pivot, где в general - общее количество заемщиков,
# в target количество должников и в ratio соотношение должников к общему числу клиентов
children_pivot = pivot_table('children')
display(children_pivot)
```

Чуть меньшая доля по несвоевременной выплате кредитов приходится на семьи с 1-м и 2-мя детьми, их соотношения примерно равны. Лучше кредитная история у самой многочисленной группы - кредитополучателей без детей, их платежеспособность выше за счет меньшей финансовой нагрузки. Не имеют задолженности по кредитам самые многодетные - с 5-ю детьми, но и число заемщиков данной категории минимальное.
